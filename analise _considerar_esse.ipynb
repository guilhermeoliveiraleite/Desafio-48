Importando a tabela e visualizando tabela : 

```python
import pandas as pd

tabela = pd.read_excel("cursinho1.xlsx")
display(tabela)
```

Media de alunos presentes : 

```python
presenca = tabela["Alunos presentes"].sum()
print(f"A media de presença é: {presenca/6:.2f} alunos por dia!")
```

A media de faltas por dia: 

```python
faltas = tabela["Alunos com falta"].sum()
print(f"A media de faltas por dia é: {faltas/6:.2f} alunos!")
```

Faltas por dia da semana: 

```python
import matplotlib.pyplot as plt 

dias = tabela["Dias da semana"]
falta = tabela["Alunos com falta"]
 

plt.title("Frequencia de aulas")
plt.pie(falta, labels=dias, autopct="%1.2f%%")
plt.show()

```

Quantidade total de alunos: 

```python
dados = {'Dias': [], 'Alunos' : []}

for x in range(6): 
    dia = tabela.iloc[x,0]
    p = tabela.iloc[x,2]
    f = tabela.iloc[x,3]

    soma = (p+f)
    soma = (f"{soma}")

    dados ['Dias'].append(dia)
    dados ['Alunos'].append(soma)

nova_tabela = pd.DataFrame(dados)
display(nova_tabela)

 
nova_tabela.to_excel('registro_alunos.xlsx', index=False)
```
