Importando a tabela e visualizando tabela : 

```python
import pandas as pd

tabela = pd.read_excel("cursinho.xlsx")
tabela2 = tabela.drop("Unnamed: 0", axis = 1)
# tabela3 = tabela2.drop('Unnamed: 1', axis = 0)

# tabela3 = tabela2.drop(['Unnamed: 1','Unnamed: 2'], axis = 0) 
# tabela.drop(0, axis=0)

display(tabela2)
```

Media de alunos presentes : 

```python
presenca = tabela2["Alunos presentes"].sum()
print(f"A media de presença é: {presenca/6:.2f} alunos por dia!")
```

A media de faltas por dia: 

```python
faltas = tabela2["Alunos com falta"].sum()
print(f"A media de faltas por dia é: {faltas/6:.2f} alunos!")
```

Faltas por dia da semana: 

```python
import matplotlib.pyplot as plt 

dias = tabela2["Dias da semana"]
falta = tabela2["Alunos com falta"]
 

plt.title("Gasto do mês por setor - Julho 2024")
plt.pie(falta, labels=dias, autopct="%1.2f%%")
plt.show()
```
